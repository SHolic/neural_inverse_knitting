```python
from PIL import Image
import os

# 指定PNG文件的路径
# img_path = '../dataset/instruction-front/2j_rayon-7.2-4-41-1.png'
img_path = '../dataset/instruction-complete/2j_rayon-7.2-4-41-1.png'
# img_path = '../dataset/instruction/Cable1_001_0_0.png'
# img_path = '../dataset-mit/instruction/Cable1_001_0_0.png'

# 打开图像
img = Image.open(img_path)

# 检查图像是否使用调色板模式（模式 'P'）
if img.mode == 'P':
    # 获取调色板
    palette = img.getpalette()
    print("调色板:", palette)
else:
    print("该图像不使用调色板模式")
```
